```python
import seaborn as sns
import matplotlib.pyplot as plt
def outlier_detector(col):
    plt.figure(figsize=(5,5))  
    sns.boxplot(x=df[col])
    plt.title(f'Box plot of {col} to detect outlier')
    plt.show()
outlier_detector('Sleep_Duration')
outlier_detector('Study_Hours')
outlier_detector('Screen_Time')
outlier_detector('Caffeine_Intake')
outlier_detector('Physical_Activity')
outlier_detector('Sleep_Quality')
outlier_detector('Weekday_Sleep_Start')
outlier_detector('Weekend_Sleep_Start')
outlier_detector('Weekday_Sleep_End')
outlier_detector('Weekend_Sleep_End')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# 1. Descriptive Statistics Heatmap
# Get the descriptive statistics for the selected columns
describe_df = df[['Sleep_Duration', 'Study_Hours', 'Sleep_Quality']].describe()

# Create heatmap for descriptive statistics
plt.figure(figsize=(10, 6))
sns.heatmap(describe_df.T, annot=True, cmap='YlGnBu', fmt='.2f', cbar=True)
plt.title('Descriptive Statistics Heatmap')
plt.tight_layout()

# Save the heatmap as an image
plt.savefig(r'C:\Users\gf\Downloads\Descriptive_Statistics_Heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# 2. Correlation Heatmap
# Calculate the correlation matrix for the selected columns
corr_matrix = df[['Sleep_Duration', 'Study_Hours', 'Sleep_Quality']].corr()

# Create heatmap for correlation
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar=True)
plt.title('Correlation Heatmap')
plt.tight_layout()

# Save the heatmap as an image
plt.savefig(r'C:\Users\gf\Downloads\Correlation_Heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure
plt.figure(figsize=(10, 6))

# Plot histogram for Sleep Duration
sns.histplot(df['Sleep_Duration'], bins=10, color='skyblue', edgecolor='black', kde=True)  # kde=True adds the line plot
plt.title('Distribution of Sleep Duration with Density Line')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')
plt.grid(True)

# Save the figure
plt.savefig(r"C:\Users\gf\Downloads\Sleep_Duration_Histogram_with_Line.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python


# Line plot for Study Hours
plt.figure(figsize=(12, 6))
plt.plot(df['Student_ID'], df['Study_Hours'], color='darkblue', marker='o', linestyle='-', markersize=4, linewidth=1)
plt.title('Study Hours for Each Student')
plt.xlabel('Student ID')
plt.ylabel('Study Hours (per week)')
plt.grid(True)

# Save the figure
plt.savefig(r"C:\Users\gf\Downloads\Study_Hours_LinePlot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Bar plot for Sleep Quality
plt.figure(figsize=(10, 6))
sns.countplot(x='Sleep_Quality', data=df, palette='viridis')
plt.title('Distribution of Sleep Quality Levels')
plt.xlabel('Sleep Quality (1=Poor to 5=Excellent)')
plt.ylabel('Frequency')

# Save the figure
plt.savefig(r"C:\Users\gf\Downloads\Sleep_Quality_BarChart.png", dpi=300, bbox_inches='tight')
plt.show()
```
